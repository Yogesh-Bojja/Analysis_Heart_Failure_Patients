### 1.Introduction: <a class = "anchor" id = "indtroudction"></a>
Every year around 17 million people die due to Cardiovascular diseases. This mainly consists of heart failures. There are various laboratory tests available through which a person can obtain some useful results. For example Creatinine phosphokinase responsible in regeneration of ATP, when in excess can lead to numerous heart related diseases. In the same way there are other components present in the body which can guide us to keep a check on the progression of heart failure. Results of these different tests and some daily habits and some other features can be used in Machine Learning to predict the survival rate of a person from Cardiovascular disease which can be unseen by a doctor.

#### 1.1.Dataset Source: <a class = "anchor" id = "data_source"></a>
Chicco, D., 2020. UCI Machine Learning Repository: Heart failure clinical records Data Set. [online] Archive.ics.uci.edu. Available at: <https://archive.ics.uci.edu/ml/datasets/Heart+failure+clinical+records> [Accessed 10 April 2021].

#### 1.2.Dataset Details: <a class = "anchor" id = "dataset_details"></a>
This dataset consists of records of heart failure patients. It has 299 records and 13 attributes in it. It is a multivariate dataset. Different clinical features are provided in this dataset. There are 105 women and 194 men in this dataset ranging between age group of 40 and 95.

#### 1.3.Dataset Features: <a class = "anchor" id = "dataset_features"></a>
| Feature Name | Data type | Units | Description |
| :- | :- | :- | :- |
| Age | Integer | years | Patient age |
| Anaemia | Integer | boolean | Reduced blood cells or hemoglobon(0: No Anaemia and 1: Anaemia) |
| High Blood Pressure | Integer | boolean | Hypertension in patient(0: No High Blood Pressure and 1: High Blood Pressure |
| Creatinine Phosphokinase (CPK) | Integer | mcg/L | CPK enzyme level in blood |
| Diabetes | Integer | boolean | Diabetes of patient(0: No Diabetes and 1: Diabetes |
| Ejection Fraction  | Integer | mcg/L | CPK enzyme level in blood |
| Platelets | Integer | kiloplatelets/L | Blood Platelets count |
| Sex | Integer | boolean | Gender of a patient(0: Women and 1: Men) |
| Serum Ceratinine | Real | mg/dL | Serum Ceratinine level in blood |
| Serum Sodium | Integer | mEq/L | Serum Sodium level in blood |
| Smoking | Integer | boolean | Does the patient smoke(0: Not Smoking and 1: Smoking) |
| Time | Integer | days | Follow up period |
| Death Event | Integer | boolean | Did the patient die in the follow up period(0: No and 1: Yes) |

#### 1.4.Target Feature: <a class = "anchor" id = "target_feature"></a>
The Death Event attribute is the target variable in our dataset. We will use this target variable to predict the number of survived patients based upon the clinical test results provided of the given above attributes. We will make our model based on the attributes that gives us the best result.
